### A. Basics of Circuit Theory (RECAP): 

#### Kirschhoff's and Ohm's Laws: 

#### Lagrange Formulation of Electric Circuits
\begin{equation}
\frac{\partial L}{\partial q_i} - \frac{d }{d t} \frac{\partial L}{\partial d \dot{q}_i} = 0, \quad i \in \{1,2,3,...,N\}
\end{equation}

Set 1:
- Position: Charge accumulated at position $x$, $Q(x)$
- Momentum: Rate of change in $Q(x)$, current pass through $x$, $I(x) = \dot{Q}(x)$.

Inductor,    $L_{inductor} = \frac{1}{2} L \dot{Q}^2 - 0 \quad \quad$ Capacitor,    $L_{capacitor} = 0-\frac{1}{2C} Q^2\quad \quad$ Resistor,    $L_{resistor} = \frac{1}{2} R \dot{Q}^2 - 0$

   
Set 2: 
- Position: Magnetic flux accumulated at position $x$, $\Phi(x)$ (less frequently used)
- Momentum: Rate of change in $\Phi(x)$, the voltage at $x$, $V(x) = \dot{\Phi}(x)$.


Inductor, $L_{inductor} = 0-\frac{1}{2L} \Phi^2\quad \quad$ Capacitor,    $L_{capacitor} = \frac{1}{2} C \dot{\Phi}^2-0 \quad \quad$ Resistor:
    $L_{resistor} = 0-\frac{1}{2R} \Phi^2$

#### Hamiltonian Formulation of Electric Circuits
Hamiltonian is the Legendre transform of the Lagrangian, it is equal to constant energy and as long as the forces are conservative.
\begin{equation}H = p\dot{q} - L(q,\dot{q},t)\end{equation}

For N pairs of variables,

How to obtain the equation of motion? Two 1st order differential equation for $q_i$ and $p_i$,
\begin{equation}\dot{p}_i = -\frac{\partial H}{\partial q_i}, \quad \quad \dot{q}_i = \frac{\partial H}{\partial p_i}, \quad \frac{\partial H}{\partial t} = -\frac{\partial L}{\partial t}\end{equation}

### B. Basics of Quantum Mechanics

#### Representations in various basis choices (Optional)
$\langle u \mid$: Bra : A vector of size $1 \times N$ (think of vector $u$ or $\vec{u}$)

$\mid v \rangle$: Ket : A space vector of size $N \times 1$ (think of vector $v^T$ or $\vec{v}$)

$\langle u \mid v \rangle $: inner product, a number (think of $\vec{u}\cdot \vec{v} \equiv u^T v$)


- "It is possible to expand these vectors in another basis". What does it mean? Think of an analogy in cartesian vectors:

1. Consider a vector $\vec{r}$, we can represent it by a set of orthonormal basis vectors $\hat{e}_{x},\hat{e}_{y},\hat{e}_{z}$. 

We have  $\hat{e}_{x} \cdot \hat{e}_{x} = 1$, $\hat{e}_{x} \cdot \hat{e}_{y} = 0$ and in general $\hat{e}_{i} \cdot \hat{e}_{j} = \delta_{ij}$ where $i,j \in \{x,y,z \}$.

It spans 3D space.

$\vec{r} = r_x \hat{e}_{x} + r_y \hat{e}_{y} +  r_z \hat{e}_{z}$. 
We obtain each component by projecting it to orthonormal basis vectors, e.g. $r_x \equiv \vec{r} \cdot \hat{e}_{x}$, $r_y \equiv \vec{r} \cdot \hat{e}_{y}$ etc.

For example, $r_x = 3, r_y= 4, r_z = 5$, 

then $\vec{r} = 3 \hat{e}_{x} + 4 \hat{e}_{y} + 5 \hat{e}_{z}$.

2. We can readily choose another basis, say x-y plane is rotated by $45^o$, then the new basis is: $\hat{e}_{\tilde{x}},\hat{e}_{\tilde{y}},\hat{e}_{\tilde{z}}$, where

\begin{equation}\hat{e}_{\tilde{x}} = \frac{1}{\sqrt{2}} \left(\hat{e}_{x} + \hat{e}_{y}\right),\quad \hat{e}_{\tilde{y}} = \frac{1}{\sqrt{2}} \left(-\hat{e}_{x} + \hat{e}_{y}\right), \quad \hat{e}_{\tilde{z}} = \hat{e}_{z}.\end{equation}

In the new basis, $\vec{r} = r_\tilde{x} \hat{e}_{\tilde{x}} + r_\tilde{y} \hat{e}_{\tilde{y}} +  r_\tilde{z} \hat{e}_{\tilde{z}}$ where

\begin{equation} r_\tilde{x} = \vec{r} \cdot \hat{e}_{\tilde{x}} = (3 \hat{e}_{x} + 4 \hat{e}_{y} + 5 \hat{e}_{z}) \cdot \frac{1}{\sqrt{2}} \left(\hat{e}_{x} + \hat{e}_{y}\right) = \frac{7}{\sqrt{2}},\end{equation}

\begin{equation} r_\tilde{y} = \vec{r} \cdot \hat{e}_{\tilde{y}} = (3 \hat{e}_{x} + 4 \hat{e}_{y} + 5 \hat{e}_{z}) \cdot \frac{1}{\sqrt{2}} \left(-\hat{e}_{x} + \hat{e}_{y}\right) = \frac{1}{\sqrt{2}}\end{equation}

\begin{equation} r_\tilde{z} = r_z.\end{equation}

Then, $\vec{r} = \frac{7}{\sqrt{2}} \hat{e}_{\tilde{x}} + \frac{1}{\sqrt{2}} \hat{e}_{\tilde{y}} +  5 \hat{e}_{\tilde{z}}$.

In Bra-Ket notation,
$\vec{r} \to \mid \vec{r} \rangle$.

In numerical examples, $\mid a \rangle$ is simply a $N \times 1$ array. 
- it is said to be "the representation of $\mid a \rangle$ in a specific basis".

Not only the "Hilbert space vectors" known as kets, but also the operators have a "representation in a given basis", they become matrices.


For example, let us assume a stationary Schrödinger equation for a finite number of solutions:
    
$$H \mid n \rangle = E_n \mid n \rangle, \quad \text{where} \quad n \in \{1,2,... ,M \}.$$

Now, we assume a (complete) basis $\{\mid n \rangle \}$ as,
\begin{equation}
\mid 1 \rangle = \begin{pmatrix}1\\0\\0\\0\\.\\.\\0\end{pmatrix}_{M \times 1}, 
\quad \mid 2 \rangle = \begin{pmatrix}0\\1\\0\\0\\.\\.\\0\end{pmatrix}_{M \times 1},  
\quad \mid 3 \rangle = \begin{pmatrix}0\\0\\1\\0\\.\\.\\0\end{pmatrix}_{M \times 1},  ... 
\quad \mid M \rangle = \begin{pmatrix}0\\0\\0\\0\\.\\.\\1\end{pmatrix}_{M \times 1}.  
\end{equation}

Moreover, because it is complete, the operators mapping acting within this space can also be expanded in this basis via the resolution of identitiy, $\mathbf{1} = \sum_n \mid n \rangle \langle n \mid $

\begin{eqnarray}
H &=& H \mathbf{1} = \sum_n H \mid n \rangle \langle n \mid = \sum_n E_n \mid n \rangle \langle n \mid , \\
&=&   E_1 \mid 1 \rangle \langle 1 \mid+ E_2 \mid 2 \rangle \langle 2 \mid + ... + E_M \mid M \rangle \langle M \mid,\\
&&\\
&=& \begin{pmatrix}
E_1 & 0 & 0 & 0 & ... &0\\
0&E_2&0&0&...&0\\
0&0&E_3&0&...&0\\
.&.&.&.&...&.\\
0&0&0&0&...&E_M
\end{pmatrix}_{M \times M}
\end{eqnarray}

#### Schrödinger Equation

\hat{H}(\vec{r} ) \Psi (\vec{r} ) &=& \left(-\frac{\hbar^2}{2 m} \vec{\nabla}^2 + V(x) \right) \Psi (\vec{r} ) = E \Psi(\vec{r} )
\end{eqnarray}

It is actually the real space basis representation of the Bra-Ket version:
\begin{equation} \hat{H} \mid \Psi \rangle = \left( \frac{\hat{p}^2 }{2 m} + \hat{V} \right)  \mid \Psi \rangle \end{equation} 

- Wavefunction: $\Psi(\vec{r} ) = \langle \vec{r} \mid \Psi \rangle$
- Operators: $-i \hbar \vec{\nabla} = \langle \vec{r} \mid \hat{p} \mid \vec{r} \rangle$



#### Tight-Binding Hamiltonian
Representation of the Schrödinger equation in a localized basis, $1 = \sum \mid \vec{r} \rangle \langle \vec{r} \mid$ where each element of the basis consists of localized and orthogonal functions.

\begin{equation} \hat{H} \mid \Psi \rangle = \left( \frac{\hat{p}^2 }{2 m} + \hat{V} \right)  \mid \Psi \rangle \end{equation} 

Inserting resolution of identitiy, $\mathbf{1} = \sum_{m} \mid m \rangle \langle m \mid$ from left and right side of the equation is equivalent to the multiplication by unity. 
\begin{eqnarray}

\end{eqnarray}

#### SSH-model: Polyacetylene, fractionalization and topology

Consider one-dimensional chain consisting of alternating bonds, e.g. the kinetic energy , hopping terms- alternating between two negative values.

The Hamiltonian is given as:

\begin{equation}
\hat{H}_{SSH} = -\sum_{m=1,2,..,N-2} t_1 \mid m \rangle  \langle m+1 \mid + t_2 \mid m +1 \rangle  \langle m+2 \mid - \mu  \sum_{m=1,2,..,N} \mid m \rangle  \langle m \mid
\end{equation}

Let us analyse the spectrum and the wavefunctions numerically.

```python
#%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from numpy import *
from scipy import *
import numpy as np
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pylab import * # for plotting commandss
import matplotlib.pyplot as plt

import schemdraw
import schemdraw.elements as elm
from schemdraw import logic

#from numpy import (real,imag,matrix,linspace,array,complex)
```

```python
t1 = 1.;   # kinetic hopping amp.
t2vec = linspace(0,2,51);  # kinetic hopping amp.
N = 52;    # length of chain N/2 is the number of unit cell 

Energy = np.zeros((len(t2vec),N),dtype='float');  # eigenvalues initialization
Psi = np.zeros((len(t2vec),N,N),dtype='complex'); # eigenvectors init.

for indt2 in arange(len(t2vec)):
    t2 = t2vec[indt2];
    hopVec = reshape(np.array([t1*np.ones((int(N/2),)),t2*np.ones((int(N/2),))],dtype='float').T,(N,))
    H = -np.diag(hopVec[0:N-1],k=1)
    H = H + H.T
    En,Psin = eig(H)
    Eorder   = argsort(En.real)
    E1 = En[Eorder].real
    Psi[indt2,:,:] = Psin[:,Eorder]
    Energy[indt2,:] = E1
```

#### Graphene: Band diagram and topology
Graphene is made of Carbon atoms arranged in a honeycomb geometry, i.e. it has an effective single orbital tight-binding description in a triangular Bravais lattice with two basis elements. The Hamiltonian is,
\begin{eqnarray}
\hat{H}_{Graphene} &=& -t \sum_{m_1,m_2} \mid m_1,m_2  \rangle  \langle m_1+1,m_2 \mid + \mid m_1,m_2  \rangle  \langle m_1,m_2+1 \mid + \mid m_1,m_2  \rangle  \langle m_1+1,m_2+1 \mid +h.c. \\&+&\epsilon_0 \sum_{m_1,m_2} \mid m_1,m_2  \rangle  \langle m_1,m_2 \mid
\end{eqnarray}

Bravais lattice vectors and the reciprocal vectors to be used for the Bloch's condition

```python
# Bravais vectors
a1 = np.array([ 0.5,sqrt(3)*0.5,0],dtype='float')
a2 = np.array([-0.5,sqrt(3)*0.5,0],dtype='float')
a3 = np.array([0,0,1.],dtype='float')

#basis vectors
d1 = np.array([0,0 ,0]/sqrt(3),dtype='float')
d2 = np.array([0,1.,0]/sqrt(3),dtype='float')

vol = np.dot(np.cross(a1,a2),a3)

# Reciprocal vectors
b1 = 2*pi/vol*np.cross(a2,a3)
b2 = 2*pi/vol*np.cross(a3,a1)

# bond vectors
rx = (a2-2*a1)/3.
ry = (a1-2*a2)/3.
rz = (a1+a2)/3.

# special points
Gam = [0.,0.]
Mpt = (2*b1+b2)/3./pi
Kpt1 = b1/2./pi
Kpt= (b1-b2)/3./pi
```

The Hamiltonian can be readily diagonalized by a discrete Fourier transformation of Bloch basis vectors, i.e. the real space vectors can be decomposed into the momentum space vectors:
\begin{eqnarray}
\mid m_1,m_2 \rangle &=& \frac{1}{\sqrt{N}}\sum_{\vec{k}}e^{i \vec{k}\cdot\vec{R}_{m_1,m_2}} \mid k_1,k_2 \rangle,\\
H &=& \sum_{\vec{k}} H_{\vec{k}},\quad
H_{\vec{k}} = \begin{pmatrix} \epsilon_0 & e^{i \vec{k}\cdot\vec{r}_{x}}+e^{i \vec{k}\cdot\vec{r}_{y}}+e^{i \vec{k}\cdot\vec{r}_{z}}\\c.c.&\epsilon_0\end{pmatrix}
\end{eqnarray}

### C. Quantum Simulation
#### Connection between the Electric Circuits and the Quantum Mechanics
In the same spirit with the Fourier Xformed Hamiltonians, it is possible to reproduce the same Hamiltonian matrix in the Fourier space. Each individual voltage or the current mode in the network is guaranteed to have the same form as the matter waves.

There are two ways of simulating an Hamiltonian matrix
1. Generating the H-matrix out of the set of Euler-Lagrange equations


2. Generating the Laplacian matrix connecting the voltage vector to the current vector.


We have already talked about the former case, therefore here we focus on the 2nd approach.


#### Simulating Schrödinger Equation as a partial differential equation


2. It is possible to simulate the SSH-model physics - tight-binding description

It has two basis elements in a unit cell and writing down the circuit Laplacian $J$ such that $J V = I$ where $V$ and $I$ are vectors labeling the voltages and the currents respectively. $J$ consists of three matrices,
- Adjacency matrix, D
- Total conductance matrix, W 
- Grounding matrix, G

$$J = W-D+G$$

Let us label each node with vector $V = (V^A_1,V^B_1,V^A_2,V^B_2,...,V_M^B)^T$ and the current passing through each capacitor as as $I = (I_1^A,I_1^B,I_2^A,I_2^B,...,I_M^B)^T$. Note that the currents leaking through the ground are not labeled.

Then, the adjacency conductance matrix is,

$$D = \begin{pmatrix}
0& j w C_1 & 0 & 0  & ... &0&0\\
j w C_1 & 0 & j w C_2&0&...&0&0\\
0&j w C_2&0&j w C_1 &...&0&0\\
.&.&.&.&...&.&.\\
0&0&0&0&...&0&j w C_2\\
0&0&0&0&...&jw C_2&0
\end{pmatrix}$$

The total conductance matrix (out of each vertex) is,
$$W = \begin{pmatrix}
j w C_1 & 0 & 0  &0& ... &0&0\\
0&j w (C_1 +C_2) & 0&0&...&0&0\\
0&0&j w (C_1 +C_2)&0 &...&0&0\\
.&.&.&.&...&.&.\\
0&0&0&0&...&j w (C_1 +C_2)&0\\
0&0&0&0&...&0&j w C_2
\end{pmatrix}$$

And finally, the grounding matrix is an identity matrix with a fixed conductance to the groundfor each vertex is,
$$G = \frac{1}{j w L} \mathbf{1}_{2 M}. $$

Note that, we will consider an infinite chain, therefore the total conductance matrix does also become constant $W = j w (C_1 +C_2) \mathbf{1}_{2 M}$.

The SSH model circuit Laplacian can written as,
$$J_{ssH} = i w C_1\begin{pmatrix}
(1+t)(1-\frac{\tilde{w}^2}{w^2}) & -t & 0  &0& ... &0&0\\
-t&(1+t)(1-\frac{\tilde{w}^2}{w^2})& -1&0&...&0&0\\
0&-1&(1+t)(1-\frac{\tilde{w}^2}{w^2})&-t &...&0&0\\
.&.&.&.&...&.&.\\
0&0&0&0&...&(1+t)(1-\frac{\tilde{w}^2}{w^2})&-1\\
0&0&0&0&...&-1&(1+t)(1-\frac{\tilde{w}^2}{w^2})
\end{pmatrix}$$

with $\tilde{w} = 2 \pi/\sqrt{L(C_1+C_2)}$.

Let us now implement it,

```python
C2 = 1.  # in units of Farads
C1 = 0.2 # in units of Farads
L = 1.   # in units of Henry

NumSites = 15

t = C2/C1                      # kapasitansların oranı
omega_0 = (L*(C1+C2))**(-0.5)  # rezonans frekansı
omega = omega_0                # rezonansta seçilmiş olan frekans

print(omega_0)
```

Beware that we choose $\tilde{w} = w$, hence the diagonals of the Laplacian matrix vanish.

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def SSH(t):
    NumSites = 12
    omega,omega_0 = 1.,1.
    E0 = (1+t)*(1-(omega_0/omega)**2)
    H1 = E0*diag(ones(NumSites*2,),0) -t*(diag(mod(arange(0,2*NumSites-1)+1,2),1)+diag(mod(arange(0,2*NumSites-1)+1,2),-1))\
    -(diag(mod(arange(0,2*NumSites-1),2),1)+diag(mod(arange(0,2*NumSites-1),2),-1))
    [En,Psin] = eig(H1)
    Eorder   = argsort(En.real)
    E1 = En[Eorder]
    Psi = Psin[:,Eorder]
    Energy = E1
    absPsi = abs(Psi)**2
    
    # plot the mid point energy and the voltage distribution
    f, ax = plt.subplots(1,2, sharey=True,figsize=(10,6))
    ax[0].scatter(0*Energy,Energy/max(Energy))
    #ax[0].set_ylabel('Admittance Im(lambda)/ max(lambda)')
    ax[0].set_ylim([-1,1])
    ax[1].stem(arange(2*NumSites),absPsi[:,NumSites]/absPsi[:,NumSites].max())
    trend = (t**(arange(2*NumSites))*mod(arange(2*NumSites)+1,2)+\
             t**((2*NumSites-arange(2*NumSites)-1))*mod(arange(2*NumSites),2))
    ax[1].plot(arange(2*NumSites),trend/max(trend))
    ax[1].set_ylabel('|Psi|**2')
    ax[1].set_ylim([-1.2,1.2])


interact(SSH,t = widgets.FloatSlider(value=0.5,min=0.02,max=1.3,step=0.02,disabled=False))
```
